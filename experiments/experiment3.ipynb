```python
import pickle

# Function to save model as pickle file
def save_model_pickle(model, path):
    with open(path, "wb") as f:
        pickle.dump(model, f)
    print("Model saved successfully as pickle file!")

# Function to load model from pickle file
def load_model_pickle(path):
    with open(path, "rb") as f:
        model = pickle.load(f)
    model.eval()
    print("Model loaded successfully from pickle file!")
    return model


# Collate function to pad variable-length sequences
def collate_fn(batch):
    features, labels = zip(*batch)
    features_padded = pad_sequence(features, batch_first=True, padding_value=0)
    labels_padded = pad_sequence(labels, batch_first=True, padding_value=0)
    return features_padded, labels_padded
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from glob import glob
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix
from vap_sound.utils.extract_mfcc_torchaudio import MFCC_EXTRACTION

MFCC_SAMPLE_RATE = 16000
FUTURE_OFFSET = 10
HOP_LENGTH = 160
N_MFCC_PER_FRAME = 40
SEQ_LENGTH = 100    
BATCH_SIZE = 32
LEARNING_RATE = 0.001
NUM_EPOCHS = 15
BATCH_FILE_SIZE = 1000


MODEL_PATH = "vap_model_3.pkl"
LOG_MODEL_EVALUATION = "../logs/model_evaluation_3.log"
LOG_MODEL_PREDICTION = "../logs/model_prediction_3.log"

# Create logs directory if missing
os.makedirs("../logs", exist_ok=True)

class ImprovedVAPModel(nn.Module):
    def __init__(self, input_dim=40, proj_dim=64, transformer_heads=4, transformer_layers=2,
                 lstm_hidden_dim=128, lstm_layers=2, output_dim=1, dropout_prob=0.3):
        super(ImprovedVAPModel, self).__init__()

        self.input_proj = nn.Linear(input_dim, proj_dim)
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=proj_dim, nhead=transformer_heads, dropout=dropout_prob),
            num_layers=transformer_layers
        )
        self.lstm = nn.LSTM(input_size=proj_dim, hidden_size=lstm_hidden_dim, num_layers=lstm_layers,
                            batch_first=True, bidirectional=True, dropout=dropout_prob)
        self.layer_norm = nn.LayerNorm(2 * lstm_hidden_dim)
        self.dropout = nn.Dropout(dropout_prob)
        self.fc = nn.Linear(2 * lstm_hidden_dim, output_dim)

    def forward(self, x):
        x = self.input_proj(x)
        x = self.transformer(x)
        x, _ = self.lstm(x)
        x = self.layer_norm(x)
        x = self.dropout(x)
        x = self.fc(x)  # (batch, seq, 1)
        return x

class VAPDataset(Dataset):
    def __init__(self, data_df, seq_length=100):
        self.seq_length = seq_length
        self.features = [torch.tensor(f, dtype=torch.float32) for f in data_df['features']]
        self.labels = [torch.tensor(l, dtype=torch.float32) for l in data_df['labels']]

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        feature, label = self.features[idx], self.labels[idx]
        if feature.shape[0] > self.seq_length:
            feature, label = feature[-self.seq_length:], label[-self.seq_length:]
        else:
            pad_len = self.seq_length - feature.shape[0]
            feature = torch.cat([torch.zeros(pad_len, feature.shape[1]), feature], dim=0)
            label = torch.cat([torch.zeros(pad_len, label.shape[1]), label], dim=0)
        return feature, label


def train_model(training_data, batch_size=16, epochs=10, seq_length=100, learning_rate=1e-3, model=None):
    if os.path.exists(MODEL_PATH):
        model = load_model_pickle(MODEL_PATH)

    train_dataset = VAPDataset(training_data, seq_length)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    criterion = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([10.0]))

    for epoch in range(epochs):
        model.train()
        train_loss, val_loss = 0.0, 0.0

        for features, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(features)

            if labels.shape[-1] == 2:
                labels = labels[:, :, 1:]  # keep class 1 (speech)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        model.eval()
        with torch.no_grad():
            for features, labels in train_loader:
                outputs = model(features)
                if labels.shape[-1] == 2:
                    labels = labels[:, :, 1:]
                loss = criterion(outputs, labels)
                val_loss += loss.item()

        print(f"Epoch {epoch+1}/{epochs}, Train Loss: {train_loss/len(train_loader):.4f}, Val Loss: {val_loss/len(train_loader):.4f}")
    save_model_pickle(model, MODEL_PATH)
    print("Training complete!")


def evaluate_model(model, test_loader, output_file, metrics_file):
    model.eval()
    all_preds, all_labels = [], []
    criterion = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([10.0]))

    with torch.no_grad(), open(output_file, "w") as f_pred, open(metrics_file, "w") as f_metrics:
        for features, labels in test_loader:
            outputs = model(features)
            if labels.shape[-1] == 2:
                labels = labels[:, :, 1:]

            min_len = min(outputs.shape[1], labels.shape[1])
            outputs = outputs[:, :min_len, :]
            labels = labels[:, :min_len, :]

            loss = criterion(outputs, labels)
            preds = (torch.sigmoid(outputs) > 0.5).float()

            all_preds.extend(preds.cpu().numpy().flatten())
            all_labels.extend(labels.cpu().numpy().flatten())

            for p, l in zip(preds, labels):
                f_pred.write(f"Predicted: {p.tolist()}\nActual: {l.tolist()}\n\n")

        acc = accuracy_score(all_labels, all_preds)
        conf_matrix = confusion_matrix(all_labels, all_preds)

        f_metrics.write(f"Test Accuracy: {acc:.4f}\n")
        f_metrics.write(f"Confusion Matrix:\n{conf_matrix}\n")

        print(f"Test Accuracy: {acc:.4f}")
        print("Confusion Matrix:")
        print(conf_matrix)

        sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.title("Confusion Matrix")
        plt.show()

# ========== MAIN TRAINING LOOP ==========
data_dir = "/Users/shanujha/Desktop/voice_activity_prediction/voice_data_mozilla/en/clips"
audio_files = glob(os.path.join(data_dir, "*.mp3"))
file_df = pd.DataFrame({"file_paths": audio_files, "processed": [0]*len(audio_files)})

mfcc_extraction = MFCC_EXTRACTION()
model1 = ImprovedVAPModel()
total = len(file_df) // BATCH_FILE_SIZE

for i in range(0, len(file_df)-BATCH_FILE_SIZE, BATCH_FILE_SIZE):
    train_files = file_df['file_paths'][i:i+BATCH_FILE_SIZE]
    training_data = mfcc_extraction.extract_mfcc_features(
        audio_files=train_files,
        target_sample_rate=MFCC_SAMPLE_RATE,
        future_offset=FUTURE_OFFSET,
        n_mfcc=N_MFCC_PER_FRAME,
        hop_length=HOP_LENGTH
    )
    train_model(training_data, batch_size=BATCH_SIZE, epochs=NUM_EPOCHS,
                seq_length=SEQ_LENGTH, learning_rate=LEARNING_RATE, model=model1)

    random_idx = np.random.randint(0, len(file_df)-BATCH_FILE_SIZE)
    eval_files = file_df['file_paths'][random_idx:random_idx+300]
    eval_data = mfcc_extraction.extract_mfcc_features(
        audio_files=eval_files,
        target_sample_rate=MFCC_SAMPLE_RATE,
        future_offset=FUTURE_OFFSET,
        n_mfcc=N_MFCC_PER_FRAME,
        hop_length=HOP_LENGTH
    )
    eval_dataset = VAPDataset(eval_data)
    eval_dataloader = DataLoader(eval_dataset, batch_size=16, shuffle=True)
    evaluate_model(model1, eval_dataloader, output_file=LOG_MODEL_EVALUATION, metrics_file=LOG_MODEL_PREDICTION)

    print(f"BATCH {i // BATCH_FILE_SIZE + 1} of {total} completed")

print("Training Completed")
```
